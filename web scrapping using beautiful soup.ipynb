```python
import requests
from bs4 import BeautifulSoup
import urllib.request
from pprint import pprint
  
url = 'https://www.searchenginejournal.com/'
  
reqs = requests.get(url)
  
soup = BeautifulSoup(reqs.text, 'html.parser')
my_data = []
  
# displaying the title
print("Title of the website is : ")
for title in soup.find_all('title'):
    print(title.get_text())

# displaying description
print("Decsription of the website is:")
page = urllib.request.urlopen('https://www.searchenginejournal.com/')  
html = BeautifulSoup(page.read(),"html.parser")
for tags in html.find_all('meta'):
    print(tags.get('content'))
    
#displaying link
#inks = soup.select_one('link[rel="image_src"]')['href']
print("Links on this websites:")
inks = soup.find_all("link", {"rel": True})
for link in links:
    print(link.attrs["href"])
```
